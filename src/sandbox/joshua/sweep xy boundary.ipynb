```julia
using Images
using PyPlot
include("/Users/joshua/Documents/GitDepot/depot/cloudlab/src/sandbox/joshua/pointcloud.jl");
filename = "/Users/joshua/Documents/Projects/elegant-prize-3149/source/po_scan/a/005/sweep.xyz";
xyz = load_xyz(filename);
```

```julia
plot(xyz[:,1],xyz[:,2]);
```

```julia
D = sqrt(sum(xyz[:,1:2].^2,2));
Theta = atan2(xyz[:,1],xyz[:,2])*180/pi;
dtheta = 0.5;
```

```julia
function get_boundary(xy)
    i = 1;
    x = minimum(Theta);
    boundary = zeros(ifloor((maximum(Theta)-minimum(Theta))/dtheta)+1,2);
    while x<maximum(Theta)
        theta_mask = x .< Theta .<= (x+dtheta);
        boundary[i,:] = xy[indmax(D.*theta_mask),:];
        i += 1;
        x += dtheta;
    end
    return boundary
end
```

```julia
@time boundary = get_boundary(xyz[:,1:2]);
```

```julia
plot(xyz[:,1],xyz[:,2],"r.");plot(boundary[:,1],boundary[:,2],".");
```
